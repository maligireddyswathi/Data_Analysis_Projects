```python
# Plot 2: Fuel Type Distribution (Pie chart)
fuel_counts = df["Fuel_Type"].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(fuel_counts, labels=fuel_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Fuel Type Distribution")
plt.show()
```

```python
# Plot 3: Transmission vs Price (Box plot)
plt.figure(figsize=(10, 5))
sns.boxplot(x=df["Transmission"], y=df["Price"])
plt.title("Transmission Type vs Price")
plt.xlabel("Transmission Type")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot 4: Yearly Car Count
plt.figure(figsize=(12, 5))
sns.countplot(x=df["Year"], order=sorted(df["Year"].unique()))
plt.title("Number of Cars per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot 5: Mileage Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df["Mileage"], bins=50, kde=True)
plt.title("Distribution of Car Mileage")
plt.xlabel("Mileage (km)")
plt.ylabel("Count")
plt.show()
```

```python
# Plot 6: Engine Size Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df["Engine_Size"], bins=30, kde=True)
plt.title("Distribution of Engine Sizes")
plt.xlabel("Engine Size (L)")
plt.ylabel("Count")
plt.show()
```

```python
# Plot 7: Doors vs Price (Box plot)
plt.figure(figsize=(10, 5))
sns.boxplot(x=df["Doors"], y=df["Price"])
plt.title("Number of Doors vs Price")
plt.xlabel("Number of Doors")
plt.ylabel("Price")
plt.show()
```

```python
# Plot 4: Year vs. Average Price (Line Plot)
avg_price_by_year = df.groupby("Year")["Price"].mean()

plt.figure(figsize=(12, 6))
sns.lineplot(x=avg_price_by_year.index, y=avg_price_by_year.values, marker="o", color="blue")
plt.title("Year vs Average Car Price")
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.grid(True)
plt.show()
```
